```python
from __future__ import print_function

import os
import sys
import pandas as pd
import numpy as np
import matplotlib as plt

from datetime import datetime
from matplotlib import dates
%matplotlib inline
```

```python
DFData=os.getenv('DFDATA')
```

```python
# this it is from MohitÂ´s notebooks

dt_parse = lambda x: datetime.strptime(x, '%m/%d/%Y')
df= pd.read_csv(DFData + "/p424-amsu/1414245891/p424-amsu",parse_dates=['Date'],date_parser=dt_parse,index_col='Date')
df.head()
print (df.head())
```

```python
df.columns
df.drop([u'ID', u'Segment ID', u'Roadway Name', u'From', u'To', u'Direction', u'12:00-1:00 AM', u'1:00-2:00AM', u'2:00-3:00AM',
       u'3:00-4:00AM', u'4:00-5:00AM', u'5:00-6:00AM', u'6:00-7:00AM',
       u'7:00-8:00AM', u'8:00-9:00AM', u'9:00-10:00AM', u'10:00-11:00AM',
       u'11:00-12:00PM', u'12:00-1:00PM', u'1:00-2:00PM',
       u'3:00-4:00PM', u'4:00-5:00PM', u'5:00-6:00PM', u'6:00-7:00PM',
       u'7:00-8:00PM', u'8:00-9:00PM', u'9:00-10:00PM', u'10:00-11:00PM',
       u'11:00-12:00AM'],axis=1, inplace = True)
```

```python
print (df.head())
```

```python
ax = df.plot(stacked=False, figsize=(12, 8), color='#3F5D7D',rot=90)

ax.set_ylabel('Number of cars',size=10)
ax.set_xlabel('Date',size=10)
ax.set_title('Number of cars at 2:00-3:00PM')
```
