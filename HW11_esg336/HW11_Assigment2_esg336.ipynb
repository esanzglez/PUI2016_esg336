# Getting NYC ZipCodes

```python
Zipcode = gp.read_file('nyc-zip-code-tabulation-areas-polygons.geojson')
Zipcode.head(3)
```

```python
Zipcode  = Zipcode.drop([ u'@id', u'BLDGpostalCode', u'PO_NAME', u'CTY_FIPS', u'ST_FIPS',
                             u'STATE', u'OBJECTID', u'Shape_Area', u'borough', u'Shape_Leng'], axis = 1)
Zipcode = Zipcode.rename(columns={'postalCode' : 'zipcode'})
Zipcode['zipcode'] = pd.to_numeric(Zipcode['zipcode'], errors = 'coerce')
Zipcode.head()
```

```python
#cleaning the data
for i in year:
    zf = zipfile.ZipFile(os.getenv("PUIDATA") + "/" + "zbp" + i + "totals.zip")
    temp = pd.read_csv(zf.open('zbp'+i+'totals.zip'.replace('.zip','.txt')))
    temp.columns = map(str.lower, temp.columns)
    temp = temp[['zip', 'est']]
    year = dt.datetime.strptime(str(i), '%y').year
    temp.rename(columns={'est': str(year), 'zip':'zipcode'}, inplace=True)
    Zipcodes = Zipcodes.merge(temp, on = 'zipcode', how = 'inner')
```

```python
# Standarized time series
ts_mean = Zipcodes.iloc[:,range(2,23)].mean(axis = 1)
ts_std = Zipcodes.iloc[:,range(2,23)].std(axis = 1)
Zipcodes.iloc[:,range(2,23)] = Zipcodes.iloc[:,range(2,23)].apply(lambda x: (x-ts_mean)/ts_std)
Zipcodes.head(3)
```

```python
print 'After standardizing the time series - \n - Std. deviation along all zip codes = {a}\
\n - Mean along all zip codes = {b}(~0)'.format(
    a = Zipcodes.iloc[:,range(2,23)].std(axis = 1).sum()/len(Zipcodes),
    b = Zipcodes.iloc[:,range(2,23)].mean(axis = 1).sum()/len(Zipcodes))
```

# Clustering the data

```python
X = Zipcodes.iloc[:, range(2,23)]
range_n_clusters = range(2,10)
for n_clusters in range_n_clusters:
    km = KMeans(n_clusters=n_clusters, random_state=324)
    cluster_labels = km.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    print("For n_clusters = {},".format(n_clusters)+" the average silhouette_score is {}".format(silhouette_avg))
```

Looking at the Silhouette score, the optimal number of clusters can be 2.

### Clustering using K-means

```python
# K-Means cluster with 2 clusters
kmeans_cluster2 = KMeans(2, random_state= 125).fit(Zipcodes.iloc[:, range(2,23)])
centers2 = kmeans_cluster2.cluster_centers_
Zipcodes['km2_labels'] = kmeans_cluster2.labels_
```
